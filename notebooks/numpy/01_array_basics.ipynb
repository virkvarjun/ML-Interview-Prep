**Interview focus:** array creation, indexing, slicing, reshaping, dtypes, and memory layout.

```python
import numpy as np
```

## 1. Array Creation

```python
# From Python lists
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]])

# Common constructors
zeros = np.zeros((3, 4))          # float64 by default
ones = np.ones((2, 3), dtype=np.float32)
arange = np.arange(0, 10, 2)      # [0, 2, 4, 6, 8]
linspace = np.linspace(0, 1, 5)   # evenly spaced
eye = np.eye(3)                   # identity matrix
rand = np.random.randn(3, 3)      # standard normal

print(f"shape: {b.shape}, ndim: {b.ndim}, dtype: {b.dtype}, size: {b.size}")
```

## 2. Indexing & Slicing

**Key interview point:** slices return *views* (shared memory), fancy indexing returns *copies*.

```python
x = np.arange(12).reshape(3, 4)
print(x)

# Basic indexing
print(x[1, 2])       # scalar at row 1, col 2
print(x[1])          # row 1 (1D slice)
print(x[:, 1])       # column 1
print(x[1:3, 0:2])   # submatrix

# Boolean indexing
mask = x > 5
print(x[mask])       # flattened selected elements

# Fancy indexing (returns copy)
rows = np.array([0, 2])
cols = np.array([1, 3])
print(x[rows, cols])  # [x[0,1], x[2,3]]
```

```python
# View vs copy demo
original = np.arange(6).reshape(2, 3)
view = original[0:2, 0:2]       # view — shares memory
view[0, 0] = 99
print("After modifying view:", original)

original2 = np.arange(6).reshape(2, 3)
copy = original2[0:2, 0:2].copy()  # explicit copy
copy[0, 0] = 99
print("After modifying copy:", original2)  # unchanged
```

## 3. Reshaping & Transposing

```python
a = np.arange(12)
b = a.reshape(3, 4)       # -1 infers dimension: reshape(3, -1)
c = b.flatten()           # copy
d = b.ravel()             # view when possible
e = b.T                     # transpose (view for 2D)

# Add/remove dimensions
f = a[np.newaxis, :]        # shape (1, 12)
g = a[:, np.newaxis]        # shape (12, 1)
h = np.squeeze(f)           # remove size-1 dims

print(b.shape, e.shape, f.shape)
```

## 4. Stacking & Splitting

```python
a = np.ones((2, 3))
b = np.zeros((2, 3))

vstack = np.vstack([a, b])    # (4, 3) — along axis 0
hstack = np.hstack([a, b])    # (2, 6) — along axis 1
concat = np.concatenate([a, b], axis=0)

split = np.split(vstack, 2, axis=0)  # list of 2 arrays
print(vstack.shape, hstack.shape)
```
