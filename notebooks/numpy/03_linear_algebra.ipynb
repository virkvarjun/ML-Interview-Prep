# NumPy: Linear Algebra

**Interview focus:** dot products, matrix multiply, norms, eigendecomposition, SVD, and solving linear systems.

```python
import numpy as np
```

## 1. Dot Product vs Element-wise Multiply

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))       # 32 — inner product
print(a @ b)              # same
print(a * b)              # [4, 10, 18] — element-wise

# Matrix multiply
A = np.arange(6).reshape(2, 3)
B = np.arange(9).reshape(3, 3)
print((A @ B).shape)      # (2, 3)
print(np.matmul(A, B).shape)
```

## 2. Implement Matrix Multiply from Scratch

```python
def matmul_naive(A, B):
  """C[i,j] = sum_k A[i,k] * B[k,j]"""
  m, k = A.shape
  k2, n = B.shape
  assert k == k2
  C = np.zeros((m, n))
  for i in range(m):
    for j in range(n):
      C[i, j] = np.dot(A[i], B[:, j])
  return C

A = np.random.randn(3, 4)
B = np.random.randn(4, 5)
assert np.allclose(matmul_naive(A, B), A @ B)
```

## 3. Norms & Distances

```python
v = np.array([3.0, 4.0])

print(np.linalg.norm(v))           # L2: 5.0
print(np.linalg.norm(v, ord=1))    # L1: 7.0
print(np.linalg.norm(v, ord=np.inf))  # L-inf: 4.0

# Frobenius norm of a matrix
M = np.random.randn(3, 4)
print(np.linalg.norm(M, 'fro'))
```

## 4. Solving Linear Systems

```python
# Solve Ax = b
A = np.array([[3, 1], [1, 2]], dtype=float)
b = np.array([9, 8], dtype=float)

x = np.linalg.solve(A, b)  # preferred over inv(A) @ b
print(f"x = {x}")
print(f"Verify: {A @ x}")
```

## 5. Eigendecomposition

```python
# For symmetric matrices, eigenvalues are real and eigenvectors are orthogonal
S = np.array([[4, 2], [2, 3]], dtype=float)
eigenvalues, eigenvectors = np.linalg.eig(S)

print(f"Eigenvalues:  {eigenvalues}")
print(f"Eigenvectors:\n{eigenvectors}")

# Verify: S @ v = lambda * v
for i in range(len(eigenvalues)):
  v = eigenvectors[:, i]
  lam = eigenvalues[i]
  residual = np.linalg.norm(S @ v - lam * v)
  print(f"  Residual for λ={lam:.2f}: {residual:.2e}")
```

## 6. SVD — Singular Value Decomposition

Every matrix A (m×n) = U Σ Vᵀ where U and V are orthogonal.

```python
A = np.random.randn(5, 3)
U, s, Vt = np.linalg.svd(A, full_matrices=False)

print(f"U: {U.shape}, s: {s.shape}, Vt: {Vt.shape}")

# Reconstruct
Sigma = np.diag(s)
A_reconstructed = U @ Sigma @ Vt
print(f"Reconstruction error: {np.linalg.norm(A - A_reconstructed):.2e}")
```

```python
# Low-rank approximation (truncated SVD) — used in PCA
rank = 2
A_approx = U[:, :rank] @ np.diag(s[:rank]) @ Vt[:rank, :]
print(f"Rank-{rank} approx error: {np.linalg.norm(A - A_approx):.4f}")
```

## 7. PCA from Scratch (using SVD)

```python
from numpy.random import default_rng

rng = default_rng(42)
X = rng.standard_normal((100, 5))
X += np.array([1, 2, 3, 4, 5])  # add mean shift

# Center the data
X_centered = X - X.mean(axis=0)

# SVD on centered data
U, s, Vt = np.linalg.svd(X_centered, full_matrices=False)

# Principal components are rows of Vt
# Project onto first 2 components
X_pca = X_centered @ Vt[:2].T
print(f"Original: {X.shape} → PCA: {X_pca.shape}")
print(f"Variance explained (top 2): {(s[:2]**2 / (s**2).sum() * 100).round(1)}%")
```

## 8. Practice Problems

```python
# Problem: Compute cosine similarity between all pairs of row vectors
X = np.random.randn(4, 3)
X_norm = X / np.linalg.norm(X, axis=1, keepdims=True)
cosine_sim = X_norm @ X_norm.T
print(cosine_sim.round(3))

# Problem: Least squares — fit y = mx + b
x_data = np.linspace(0, 10, 20)
y_data = 2.5 * x_data + 7 + np.random.randn(20) * 2

A = np.vstack([x_data, np.ones(len(x_data))]).T  # (n, 2)
m, b = np.linalg.lstsq(A, y_data, rcond=None)[0]
print(f"Fit: y = {m:.2f}x + {b:.2f}")
```
